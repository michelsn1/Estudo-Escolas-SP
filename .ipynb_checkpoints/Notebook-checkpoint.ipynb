```python
import pandas as pd
import numpy as np

dados=pd.read_csv('Dados - Final - Final.csv',decimal=',')
dados.head()
```

```python
from bokeh.models import ColumnDataSource, LabelSet
from bokeh.plotting import figure, show, output_file
from bokeh.sampledata.periodic_table import elements

palette = ["#053061", "#2166ac", "#4393c3", "#92c5de", "#d1e5f0",
           "#f7f7f7", "#fddbc7", "#f4a582", "#d6604d", "#b2182b", "#67001f"]

x_1= "Idade Escola"
y_1="Escolas Totais"
z_1="IDH"

idade_escola = dados[z_1]
low = min(idade_escola)
high = max(idade_escola)
idade_escola_inds = [int(10*(x-low)/(high-low)) for x in idade_escola] #gives items in colors a value from 0-10
dados['colors'] = [palette [::-1][i] for i in idade_escola_inds]


TITLE = "IDH x Número de Escolas Públicas - Distritos São Paulo"
p = figure(plot_width=1200, title=TITLE)
p.xaxis.axis_label = "IDH"
p.yaxis.axis_label = "Escolas Totais"


source = ColumnDataSource(dados)

p.circle(x_1, y_1, size=12, source=source,
         color='colors', line_color="black", fill_alpha=0.8)


labels = LabelSet(x=x_1, y=y_1, text="Distrito", y_offset=8,
                  text_font_size="8pt", text_color="#555555",
                  source=source, text_align='center')

p.add_layout(labels)


show(p)
```
